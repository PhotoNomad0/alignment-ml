```python
def describeLemmas(lemmasPath, descrPath):
    lemma_data = file.readJsonFile(lemmasPath)
    lemmas = list(lemma_data.keys())
    lemmas.sort()
    data = {}
    for lemma in lemmas:
        print(f"Describing {lemma}")
        lemmaAlignments = db.loadAlignmentDataFromFile(lemma)
        if not lemmaAlignments is None:
            descr = db.describeAlignments(lemmaAlignments, silent = True)
            descr['alignments_count'] = len(lemmaAlignments)
            data[lemma] = descr
        else:
            print(f"Skipping {lemma} since file not found")

    print(f"Described {len(lemmas)} lemmas: {lemmas}")
    file.writeJsonFile(descrPath, data)

lemmasPath = './data/kt_en_NT_lemmas.json'
descrPath = './data/kt_en_NT_descr.json'

start = time.time()
describeLemmas(lemmasPath, descrPath)
delta = (time.time() - start)
elapsed = str(timedelta(seconds=delta))
print(f'Get kt Descriptions, Elapsed time: {elapsed}')
```
